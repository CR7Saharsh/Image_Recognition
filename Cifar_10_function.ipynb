```python
#Importing necessary libraries
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.densenet import preprocess_input
from tensorflow.keras.applications.densenet import decode_predictions
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras.datasets import cifar10
```

```python
#Using our presaved model
model_path = '/content/cifar10dense101.h5'
model = load_model(model_path)
```

```python
def custom_cifar10_prediction(image_path, model):
    try:
        # Load and preprocess the user-provided image
        image = load_img(image_path, target_size=(32, 32))
        image = img_to_array(image)
        image = np.expand_dims(image, axis=0)
        image = preprocess_input(image)

        # Make predictions using the model
        predictions = model.predict(image)

        # Convert the predicted class index to class name
        class_index = np.argmax(predictions)
        class_name = cifar10_classes[class_index]

        # Load and display the input image
        img = mpimg.imread(image_path)

        # Create a figure and axes for plotting
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

        # Display the input image without grid lines
        ax1.imshow(img)
        ax1.set_title(f"Input Image")

        # Display the input image without grid lines and the predicted class label
        ax2.imshow(img)
        ax2.set_title(f"Predicted class: {class_name}")

        plt.show()

    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Define the CIFAR-10 class names (you can define this based on your model)
cifar10_classes = [
    'Airplane',
    'Automobile',
    'Bird',
    'Cat',
    'Deer',
    'Dog',
    'Frog',
    'Horse',
    'Ship',
    'Truck'
]

# Define the CIFAR-10 class names (you can define this based on your model)
cifar10_classes = [
    'Airplane',
    'Automobile',
    'Bird',
    'Cat',
    'Deer',
    'Dog',
    'Frog',
    'Horse',
    'Ship',
    'Truck'
]
```

```python
image_path = '/content/Horse-Facebook.jpg'

# Call the custom prediction function
custom_cifar10_prediction(image_path,model)
```
